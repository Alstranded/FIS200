## Gráficos Rapidez

```python
step = 4

# velocidad a
v_a = []
# tiempo velocidad a
t_v_a = []
v_m = []
t_v_m = []
v_b = []
t_v_b = []

cant_data_a = len(t_a)
cant_data_m = len(t_m)
cant_data_b = len(t_b)

for i in range(cant_data_a - step):
    v = (x_a[i+step] - x_a[i]) / (t_a[i+step] - t_a[i])
    t = t_a[i] + (t_a[i+step] - t_a[i])/2
    v_a.append(v)
    t_v_a.append(t)

for i in range(cant_data_m - step):
    v = (x_m[i+step] - x_m[i]) / (t_m[i+step] - t_m[i])
    t = t_m[i] + (t_m[i+step] - t_m[i])/2
    v_m.append(v)
    t_v_m.append(t)

for i in range(cant_data_b - step):
    v = (x_b[i+step] - x_b[i]) / (t_b[i+step] - t_b[i])
    t = t_b[i] + (t_b[i+step] - t_b[i])/2
    v_b.append(v)
    t_v_b.append(t)
```

## Ajustando una curva para la Posición

Dado una variable `pol` tipo Numpy.poly1d (polinomio), y un número `x`:

- Para evaluar un polinomio debe hacer: `pol(x)`.
- Para obtener los coeficientes del polinomio basta con: `pol.c`.

```python
grado = 2
# coeficiente a
coef_a = np.polyfit(t_a, x_a, grado)
# polinomio a
pol_a = np.poly1d(coef_a)
print(pol_a)

coef_m = np.polyfit(t_m, x_m, grado)
pol_m = np.poly1d(coef_m)
print(pol_m)

coef_b = np.polyfit(t_b, x_b, grado)
pol_b = np.poly1d(coef_b)
print(pol_b)
```

### Rapidez como x'

```python
# polinomio a prima
pol_a_p = pol_a.deriv()
print(pol_a_p)

pol_m_p = pol_m.deriv()
print(pol_m_p)

pol_b_p = pol_b.deriv()
print(pol_b_p)
```

### Aceleración como x''

```python
# polinomio a prima prima
pol_a_pp = pol_a_p.deriv()
print(pol_a_pp)

pol_m_pp = pol_m_p.deriv()
print(pol_m_pp)

pol_b_pp = pol_b_p.deriv()
print(pol_b_pp)
```

## Tiempo en Función de la Rapidez Inicial

Genere un modelo que le permita predecir el tiempo que el objeto demora en detenerse a partir de la velocidad inicial del movimiento

Para hacer el modelo se utilizará la rapidez máxima de los datos como rapidez inicial.

```python
print("Rapidez A:")
print(np.max(v_a))
print(t_a[len(t_a)-1])
print("\nRapidez B:")
print(np.max(v_m))
print(t_m[len(t_m)-1])
print("\nRapidez C:")
print(np.max(v_b))
print(t_b[len(t_b)-1])
```

### Graficando los Datos

```python
x_modelo = [np.max(v_b), np.max(v_m), np.max(v_a)]
y_modelo = [t_b[len(t_b)-1], t_m[len(t_m)-1], t_a[len(t_a)-1]]
```

```python
s = 0.8
fig, ax = plt.subplots(figsize=(10*s,8*s))
ax.plot(x_modelo, y_modelo, ".")
ax.plot(x_modelo, y_modelo, "-")
ax.set_xlabel("Rapidez Inicial [cm/s]")
ax.set_ylabel("Tiempo final [s]")
ax.set_title("Tiempo Final vs Rapidez Inicial")

plt.show()
```

### Ajuste y Gráfico

```python
grado = 1
# coeficiente modelo
coef_modelo = np.polyfit(x_modelo, y_modelo, grado)
# polinomio modelo
pol_modelo = np.poly1d(coef_modelo)
print(pol_modelo)
```

```python
linspace_length = 300

# x inicial modelo
x_i_modelo= x_modelo[0]
# x final modelo
x_f_modelo = x_modelo[len(x_modelo)-1]
#tiempo ajuste a
x_aj_modelo = np.linspace(x_i_modelo, x_f_modelo, linspace_length)
```

```python
s = 1.2
fig, ax = plt.subplots(figsize=(12*s,7*s))
ax.plot(x_modelo, y_modelo, ".")
ax.plot(x_aj_modelo, pol_modelo(x_aj_modelo), "-")
ax.set_xlabel("Rapidez Inicial [cm/s]", fontsize=15*s)
ax.set_ylabel("Tiempo final [s]", fontsize=15*s)
#ax.set_title("Tiempo Final vs Rapidez Inicial", fontsize=18*s)
ax.legend(["Datos", "Ajuste"], fontsize=13*s)
ax.tick_params(axis='both', labelsize=14*s)
plt.show()
```
